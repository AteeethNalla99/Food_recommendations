```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
df2['cuisines'] = df2['cuisines'].fillna('').astype(str).str.lower().str.strip()

# Use split on ',' without assuming a space
df2['cuisines'] = df2['cuisines'].str.split(',')

# Strip spaces from each cuisine and remove empties
df2['cuisines'] = df2['cuisines'].apply(lambda lst: [c.strip() for c in lst if c.strip()])

# Now encode
mlb = MultiLabelBinarizer()
cuisine_encoded = pd.DataFrame(mlb.fit_transform(df2['cuisines']), columns=mlb.classes_)

# Your dummies for location/rest_type
df_encoded = pd.get_dummies(
    df2.drop(columns=['cuisines']),
    columns=['location', 'rest_type'],
    drop_first=True,
    dtype=int
)

# Merge back cuisine one-hot columns
df_encoded = pd.concat([df_encoded, cuisine_encoded], axis=1)
```

```python
df_encoded.head()
```

```python
def recommend_restaurants(location_input, online_order_input, book_table_input, cuisines_input, top_n=5):
    def normalize(name):
        return name.lower().replace('_', ' ').strip()

    # 1️⃣ Location column check
    possible_location_cols = [
        col for col in df_encoded.columns
        if col.lower() == f"location_{location_input.lower().replace(' ', '_')}"
    ]
    if not possible_location_cols:
        print(f"No restaurants found in location: {location_input}")
        return []
    location_col = possible_location_cols[0]

    # 2️⃣ Filter by location
    filtered_df = df_encoded[df_encoded[location_col] == 1]

    # 3️⃣ Filter by online_order & book_table
    filtered_df = filtered_df[
        (filtered_df['online_order'] == online_order_input) &
        (filtered_df['book_table'] == book_table_input)
    ]

    # 4️⃣ Cuisine matching (case-insensitive)
    cuisines_lookup = {normalize(col): col for col in df_encoded.columns}
    matched_cols = [cuisines_lookup[normalize(c)] for c in cuisines_input if normalize(c) in cuisines_lookup]

    # If cuisine doesn't exist as a column → no such cuisine in dataset at all
    if not matched_cols:
        print("There is no cuisine like this in this restaurant.")
        return []

    # Apply cuisine filter within filtered dataset
    filtered_df = filtered_df[filtered_df[matched_cols].sum(axis=1) > 0]

    # If after filtering there are no rows → cuisine exists in dataset but not in this restaurant/location
    if filtered_df.empty:
        print("There is no cuisine like this in this restaurant.")
        return []

    # 5️⃣ Sort & return
    filtered_df = filtered_df.sort_values(by=['rate', 'votes'], ascending=False)
    return filtered_df['name'].head(top_n).tolist()
```

```python
print(recommend_restaurants("Banashankari", 1, 0, ['north indian']))
```
